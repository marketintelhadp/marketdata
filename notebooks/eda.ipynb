```python
import os

def list_files(root_dir, extensions=('.csv', '.h5')):
    result = {ext: [] for ext in extensions}
    for root, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(extensions):
                ext = os.path.splitext(file)[1]
                full_path = os.path.join(root, file)
                result[ext].append(full_path.replace("\\", "/"))  # Normalize for web use
    return result

# Set base directories
dataset_dir = 'D:\Git Projects\Price_forecasting_project\Agricultural-Price-Intelligence\data\raw\processed'
model_dir = 'D:\Git Projects\Price_forecasting_project\Agricultural-Price-Intelligence\models'

# Get file listings
datasets = list_files(dataset_dir, extensions=('.csv',))
models = list_files(model_dir, extensions=('.h5',))

# Output results
print("\nüìÅ Datasets Found:")
for path in datasets['.csv']:
    print(f" - {path}")

print("\nüìÅ Models Found:")
for path in models['.h5']:
    print(f" - {path}")
```

```python
import os

def list_all_datasets(base_dir):
    dataset_paths = []

    # Traverse base_dir and subdirectories
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.endswith(('.csv', '.xlsx')):
                full_path = os.path.join(root, file)
                dataset_paths.append(full_path)

    return dataset_paths

# Use raw string to avoid backslash escape errors
data_dir = r"D:\Git Projects\Price_forecasting_project\Agricultural-Price-Intelligence\data\raw\processed"
datasets = list_all_datasets(data_dir)

print("üìÇ Found Dataset Files:")
for path in datasets:
    print(" -", path)
```

```python
# Create a copy of the dataset to avoid modifying the original
df_continuous = df.copy()

# Assign sequential numbering for each Variety (and optionally Grade)
df_continuous['Sequence'] = df_continuous.groupby(['Variety', 'Grade']).cumcount() + 1

# Plot for each Variety and Grade combination
plt.figure(figsize=(14, 8))

# Group the data by Variety and Grade
grouped = df_continuous.groupby(['Variety', 'Grade'])

for (variety, grade), subset in grouped:
    if not subset.empty:  # Ensure the subset isn't empty
        plt.plot(
            subset['Sequence'], 
            subset['Avg Price (per kg)'], 
            label=f"{variety} - {grade}", 
            linestyle='-'
        )

# Add legend, labels, and title
plt.legend(loc='best', fontsize=8)
plt.xlabel('Observation Number')
plt.ylabel('Average Price')
plt.title('Continuous Trends by Variety and Grade (Using Sequential Numbering)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
case_weights = {
    'FC': 16,  # FC case type has a weight of 16 kg
    'HC': 8    # HC case type has a weight of 8 kg
}

# Function to convert price to per kg
def price_per_kg(row):
    case_type = row['CaseType']
    
    # Handle missing values
    if pd.isna(case_type):
        return None, None, None  # Return None for missing values
    
    # Retrieve weight based on CaseType
    weight = case_weights.get(case_type)
    
    if weight is None:  # Handle cases where case type is not in the dictionary
        return None, None, None
    
    # Calculate per kg prices
    return row['Min Price'] / weight, row['Max Price'] / weight, row['Avg Price'] / weight

# Apply the function to calculate per kg prices
df[['Min Price (per kg)', 'Max Price (per kg)', 'Avg Price (per kg)']] = df.apply(
    lambda row: pd.Series(price_per_kg(row)), axis=1
)

# Show the DataFrame with the calculated per kg prices
print(df)
```
